```python
# Filtering of the world population for obtaining Spanish population
sp=wp[ wp['Location'] == 'Spain']
# Get the set of starting ages 
age_starts=numpy.unique( sp['AgeGrpStart'] )
print( age_starts )
# Get the set of years
years=numpy.unique( sp['Year'] )
print( years )
# Get the set of age groups -- numpy.unique() do not returns a sorted list according to our needs
age_groups=list()
for age_group in sp['AgeGrp']:
    if age_group not in age_groups:
            age_groups.append( age_group )
print( age_groups )
```

```python
# Prepare the color palette from existing ones.

#colors = brewer[ 'Paired' ][ 12 ]
#colors += brewer[ 'Dark2'][8]
#colors += brewer[ 'Accent'][8]

import bokeh.palettes

colors = bokeh.palettes.magma( len(age_starts) )
#colors = bokeh.palettes.inferno( len(age_starts) )
#colors = bokeh.palettes.plasma( len(age_starts) )
#colors = bokeh.palettes.viridis( len(age_starts) )

print(colors)
```

```python
fig=figure( title="Evolution of Spanish population (1950-2100)", plot_height=400, plot_width=900 )

for i in range(len(age_starts)):
    if 0 <= age_starts[i] <= 80:
        fig.line( years, data[i], color=colors[i], line_width=2, legend=age_groups[i] )
show(fig)
```
